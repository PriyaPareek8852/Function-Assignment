Q.1.  What is the difference between a function and a method in Python?

Ans. A function is a block of reusable code that can be called anywhere in a program to perform a specific task. A method, on the other hand, is a function that's associated with an object and is called on an instance of that object.

In Python, the fundamental difference between a function and a method is that a method is a function that belongs to an object. They are essentially the same concept, but their context of use is what distinguishes them. A method must be called on an object, while a function can be called on its own

Function example:

```python
# A simple function
def greet(name):
    return f"Hello, {name}!"

# Calling the function
print(greet("World"))
```

Method Example:

```python
# A class with a method
class Dog:
    def __init__(self, name):
        self.name = name

    # This is a method
    def bark(self):
        return f"{self.name} says woof!"

# Creating an object (an instance of the Dog class)
my_dog = Dog("Buddy")

# Calling the method on the object
print(my_dog.bark())
```

Q.6. Explain the concept of generators in Python and how they are defined.



Ans. Generators are a special type of function in Python that allow you to create iterators in a simple and clean way. They are memory-efficient because they don't store all the values in memory at once. Instead, they produce values one at a time, "on the fly," as you iterate through them.

FOR EXAMPLE:

```python
def simple_generator():
    print("Starting generator...")
    yield 1
    print("Continuing...")
    yield 2
    print("Finishing...")
    yield 3

# Create an instance of the generator
gen = simple_generator()

# Iterate through the generator
print(next(gen)) # Output: Starting generator... 1
print(next(gen)) # Output: Continuing... 2
print(next(gen)) # Output: Finishing... 3
```

Q.7. What are the advantages of using generators over regular functions?



Ans. The primary advantages of using generators over regular functions are memory efficiency and lazy evaluation. Generators produce values one at a time as they are requested, which is especially useful when dealing with large datasets or infinite sequences.

Let's illustrate the difference with a practical example: generating the first n even numbers.

 Generators Function :

```python
def get_even_numbers_gen(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

# The generator is created, but no numbers are yet in memory
even_gen = get_even_numbers_gen(10000000)
# Numbers are generated and processed one by one
for number in even_gen:
    # process the number
    pass
```

Regular Function:

```python
def get_even_numbers_list(n):
    even_numbers = []
    for i in range(n):
        if i % 2 == 0:
            even_numbers.append(i)
    return even_numbers

# The entire list is created in memory
all_evens = get_even_numbers_list(10000000)
# We can now iterate over the list
for number in all_evens:
    # process the number
    pass
```

```python
def sum_even_numbers(numbers):
  """
  This function takes a list of numbers as input and returns the sum of all even numbers in the list.
  """
  even_sum = 0
  for number in numbers:
    if number % 2 == 0:
      even_sum += number
  return even_sum

my_list=[1,2,3,4,5,6,]

result=sum_even_numbers(my_list)
print(result)
```

Q.4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
  """
  Checks if a given number is prime.
  """
  if num <= 1:
    return False # Numbers less than or equal to 1 are not prime
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False # Found a divisor, so not prime
  return True # No divisors found, so prime

# Check for prime numbers from 1 to 200
print("Prime numbers between 1 and 200:")
for number in range(1, 201):
  if is_prime(number):
    print(number)
```

Q.5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
  """
  An iterator class that generates the Fibonacci sequence up to a specified number of terms.
  """
  def __init__(self, num_terms):
    self.num_terms = num_terms
    self.a = 0
    self.b = 1
    self.count = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count < self.num_terms:
      if self.count == 0:
        self.count += 1
        return self.a
      elif self.count == 1:
        self.count += 1
        return self.b
      else:
        next_fib = self.a + self.b
        self.a = self.b
        self.b = next_fib
        self.count += 1
        return next_fib
    else:
      raise StopIteration

# Example usage:
fib = FibonacciIterator(10)

for num in fib:
    print(num)
```

Q.6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
  """
  Generator function that yields powers of 2 up to a given exponent.
  """
  for exponent in range(max_exponent + 1):
    yield 2 ** exponent

# Example usage:
for power in powers_of_two(5):
  print(power)
```

Q.7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
# create a sample file
with open("sample.txt", "w") as f:
  f.write("hello\n")
  f.write("this is priya\n")
  f.write("learning python\n")

# generateor function to read file line by line
def read_file_line_by_line(filename):
  with open(filename,"r") as file:
    for line in file:
      yield line.strip() #strip removes\n at the end

# use the generator
for line in read_file_line_by_line("sample.txt"):
  print(line)
```

Q.10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(text):
  """
  Removes all vowels from a given string using the filter() function.
  """
  vowels = "aeiouAEIOU"
  # Use filter and a lambda function to keep only non-vowel characters
  filtered_text = filter(lambda char: char not in vowels, text)
  # Join the filtered characters back into a string
  return "".join(filtered_text)

# Example usage:
my_string = "Hello World"
string_without_vowels = remove_vowels(my_string)
print(f"Original string: {my_string}")
print(f"String without vowels: {string_without_vowels}")
```
